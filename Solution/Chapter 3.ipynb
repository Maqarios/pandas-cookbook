```python
# 3

# Commands
%matplotlib inline

# Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Setting Plot's Confugrations
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 5)

# Reading CSV File(s)
df_311_service_requests = pd.read_csv('../data/311-service-requests.csv', low_memory=False)
```

```python
# 3.3
# Displaying Value Counts Of Column After Filtering The Data Frame
is_noise = df_311_service_requests['Complaint Type'] == 'Noise - Street/Sidewalk'
df_noise_311_service_requests = df_311_service_requests[is_noise]
df_noise_311_service_requests['Borough'].value_counts()
```

```python
# 3.3
# Displaying Ratio Of Filtered Column(s)' Value Counts To Total Value Counts
is_noise = df_311_service_requests['Complaint Type'] == 'Noise - Street/Sidewalk'
df_noise_311_service_requests = df_311_service_requests[is_noise]
df_noise_311_service_requests['Borough'].value_counts() / df_311_service_requests['Borough'].value_counts()
# In Case Of Division By Zero In Python 1.x 2.x Add .astype(float) To The Denominator
```

```python
# 3.3
# Plotting Ratio Of Filtered Column(s)' Value Counts To Total Value Counts
is_noise = df_311_service_requests['Complaint Type'] == 'Noise - Street/Sidewalk'
df_noise_311_service_requests = df_311_service_requests[is_noise]
(df_noise_311_service_requests['Borough'].value_counts() / df_311_service_requests['Borough'].value_counts()).plot(kind='bar')
```
