```python
# 8

# Commands
%matplotlib inline

# Imports
import pandas as pd

# Reading CSV File(s) Without Last Row & Selecting Specific Columns
df_popularity_contest = pd.read_csv('../data/popularity-contest', sep=' ', )[:-1]
df_popularity_contest.columns = ['atime', 'ctime', 'package-name', 'mru-program', 'tag']
```

```python
# 8.0
# Casting Some Columns
print(df_popularity_contest.dtypes)
df_popularity_contest['atime'] = df_popularity_contest['atime'].astype(int)
df_popularity_contest['ctime'] = df_popularity_contest['ctime'].astype(int)
print(df_popularity_contest.dtypes)
df_popularity_contest['atime'] = pd.to_datetime(df_popularity_contest['atime'], unit='s')
df_popularity_contest['ctime'] = pd.to_datetime(df_popularity_contest['ctime'], unit='s')
print(df_popularity_contest.dtypes)
```

```python
# 8.0
# Applying Filter On Data Frame
df_popularity_contest = df_popularity_contest[df_popularity_contest['atime'] > '1970-01-01']
```

```python
# 8.0
# Displaying Top 10 Rows Of Filtered Data Frame
df_popularity_contest[~df_popularity_contest['package-name'].str.contains('lib')].sort_values('ctime',ascending=False)[:10]
```
