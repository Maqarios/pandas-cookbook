```python
# 7.2
# Rereading CSV File While Fixing Some Column's Values & Types Then Displaying The New Loaded Data Frame
na_values = ['NO CLUE', 'N/A', '0']
df_311_service_requests = pd.read_csv('../data/311-service-requests.csv', na_values=na_values, dtype={'Incident Zip': str})
df_311_service_requests['Incident Zip'].unique()
```

```python
# 7.3
# Slicing All Values In Some Column In A Data Frame & Displaying The Length Of A Filtered Column
df_311_service_requests['Incident Zip'] = df_311_service_requests['Incident Zip'].str.slice(0,5)
len(df_311_service_requests[df_311_service_requests['Incident Zip'].str.contains('-').fillna(False)])
```

```python
# 7.3
# Deleting Rows After Applying Some Filter
df_311_service_requests[df_311_service_requests['Incident Zip'] == '00000'] = np.nan
len(df_311_service_requests[df_311_service_requests['Incident Zip'].str.contains('-').fillna(False)])
```
