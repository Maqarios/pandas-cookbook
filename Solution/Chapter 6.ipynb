```python
# 6

# Commands
%matplotlib inline

# Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Setting Plot's Confugrations
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 3)
plt.rcParams['font.family'] = 'sans-serif'

# Reading CSV File(s)
df_weather_2012 = pd.read_csv('../data/weather_2012.csv', parse_dates=True, index_col='Date/Time')
```

```python
# 6.1
# Displaying Filtered Data Frame Of Booleans
df_weather_2012_description = df_weather_2012['Weather']
is_snowing = df_weather_2012_description.str.contains('Snow')
is_snowing.head()
```

```python
# 6.1
# Plotting Filtered Data Frame Of Booleans
df_is_snowing = is_snowing.to_frame()
df_is_snowing['Weather'].astype(int).plot()
```

```python
# 6.2
# Plotting Median Value In Some Time
df_weather_2012['Temp (C)'].resample('M').apply(np.median).plot(kind='bar')
```

```python
# 6.2
# Plotting Mean Value In Some Time
df_is_snowing.astype(int).resample('M').apply(np.mean).plot(kind='bar')
```

```python
# 6.3
# Setting 2 Data Frames' Names & Plotting 2 Data Frames Together
temperature = df_weather_2012['Temp (C)'].resample('M').apply(np.median)
snowiness = df_is_snowing.astype(int).resample('M').apply(np.mean)
temperature.name = 'Temperature'
snowiness.name = 'Snowiness'
pd.concat([temperature, snowiness], axis=1).plot(kind='Bar')
```

```python
# 6.3
# Plotting 2 Data Columns Within 1 Data Frame Seperately
pd.concat([temperature, snowiness], axis=1).plot(kind='Bar', subplots=True, figsize=(15,7))
```
